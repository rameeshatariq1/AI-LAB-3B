```python
import pandas as pd
df = pd.read_csv("data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['SepalLengthCm'] = df['SepalLengthCm'].fillna(df['SepalLengthCm'].mean())
df['SepalWidthCm'] = df['SepalWidthCm'].fillna(df['SepalWidthCm'].median())
```

```python
df['SepalLengthCm'] = df['SepalLengthCm'].astype(int)
df['SepalWidthCm'] = df['SepalWidthCm'].astype(int)
df['PetalLengthCm'] = df['PetalLengthCm'].astype(int)
df['PetalWidthCm'] = df['PetalWidthCm'].astype(int)
```
